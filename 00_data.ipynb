```python
import pandas as pd
import json
```

```python
data = []

with open("data/case_with_all_sources_with_companion_cases_tag.jsonl", "r") as file:
    for line in file:
        data.append(json.loads(line))

df = pd.DataFrame(data)
```

```python
df = df[(df["year"] >= 2010) & (df["year"] <= 2015)].reset_index(drop=True)
```
